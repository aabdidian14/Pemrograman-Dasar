```python
#Nomor 1

suhu = float(input("Masukkan suhu air pendingin (°C): "))

if suhu <= 90:
    print("STATUS: Normal")
elif suhu < 100:
    print("STATUS: Warning")
else:
    print("ALERT: SHUTDOWN")
```

Program itu dipakai untuk mengecek suhu air pendingin dan menentukan status kondisi mesin berdasarkan angka suhu yang dimasukkan.


1. suhu = float(input(...))
Program minta pengguna mengetik suhu, lalu mengubahnya menjadi angka desimal. Jadi kalau kamu ketik “92.5”, program bisa mengolahnya sebagai angka, bukan teks.


2. Bagian if suhu <= 90:
Kalau suhu masih 90°C atau lebih rendah, artinya kondisi masih aman. Karena itu program menampilkan "STATUS: Normal".


3. Bagian elif suhu < 100:
Bagian ini berjalan kalau suhu sudah lebih dari 90, tapi masih di bawah 100. Ini zona waspada, jadi program menampilkan "STATUS: Warning".


4. Bagian else:
Kalau suhunya sudah 100°C atau lebih, itu kondisi berbahaya. Program menunjukkan "ALERT: SHUTDOWN" sebagai tanda bahwa sistem seharusnya berhenti untuk mencegah kerusakan.


Jadi alurnya sederhana, seperti pengecekan bertingkat. Program memutuskan kondisi berdasarkan rentang suhu yang masuk. Tujuannya supaya operator bisa tahu kapan mesin aman, kapan waspada, dan kapan harus segera dihentikan.

```python
# Nomor 3

vibrasi_mesin = [1.2, 0.9, 2.5, 3.1, 1.8]

for i in range(len(vibrasi_mesin)):
    if vibrasi_mesin[i] > 2.0:
        print(f"Warning: Unit {i+1} vibrasi tinggi!")
```

Loop itu bekerja cukup simpel. Kamu punya list berisi lima angka vibrasi. Program melakukan pengecekan satu per satu.

1. range(len(vibrasi_mesin)) menghasilkan urutan indeks 0 sampai 4.

2. Pada setiap putaran, program melihat nilai vibrasi pada indeks tersebut.

3. Jika angkanya lebih besar dari 2.0, program mencetak peringatan untuk unit itu.

4. i+1 dipakai supaya nomor unit mulai dari 1, bukan 0.

Dengan data yang kamu punya, nilai yang melewati batas adalah 2.5 pada unit 3 dan 3.1 pada unit 4. Jadi outputnya memperingatkan dua unit itu saja. Pendekatan ini membantu teknisi cepat tahu mesin mana yang butuh pengecekan lebih lanjut.

```python
# Nomor 5

suhu_saat_ini = 60

while suhu_saat_ini < 85:
    print("Suhu saat ini:", suhu_saat_ini)
    suhu_saat_ini += 5
```

Loop akan terus jalan selama suhu masih di bawah 85°C. Setiap putaran, program nge-print suhu sekarang, lalu menaikkan 5 derajat. Begitu suhu mencapai atau melewati 85, loop berhenti otomatis.

```python
# Nomor 7

kapasitas_maks = float(input("Masukkan kapasitas maksimum (kW): "))
beban_terpakai = float(input("Masukkan beban terpakai (kW): "))

persentase = (beban_terpakai / kapasitas_maks) * 100

if persentase <= 60:
    status = "Beban Ringan"
elif persentase <= 85:
    status = "Beban Sedang"
else:
    status = "Beban Berat"

print(f"Persentase pemakaian: {persentase:.2f}%")
print(f"Status beban: {status}")
```

Program hitung dulu persentase pemakaian dari beban terhadap kapasitas maksimum. Setelah itu program cek persentasenya. Kalau masih di bawah atau sama dengan 60, dianggap ringan. Kalau lewat 60 sampai 85, itu sedang. Kalau lewat 85, masuk kategori berat. Sistem seperti ini sering dipakai untuk pantau loading generator atau panel distribusi supaya tahu kondisi operasinya.

```python
# Nomor 9

suhu_boiler = [92.5, 88.7, 95.1, 102.3, 97.8]

for i in range(len(suhu_boiler)):
    print(f"Suhu titik {i+1}: {suhu_boiler[i]}°C")

    if suhu_boiler[i] > 100:
        print("Peringatan: Suhu melewati 100°C!")
```

Loop itu muter dari titik 1 sampai titik 5. Setiap putaran, program nge-print suhu di titik tersebut. Kalau ketemu angka lewat 100°C, program langsung kasih warning karena itu sudah masuk zona rawan untuk pressure dan potensi overheating. Jadi kamu bisa cepat tahu titik mana yang butuh dicek dulu, tanpa lihat datanya satu per satu.

```python
# Challenge


def hitung_efisiensi(input_energi, output_energi):
    if input_energi == 0:
        return 0
    return output_energi / input_energi



def klasifikasi_beban(kapasitas, terpakai):
    persentase = (terpakai / kapasitas) * 100

    if persentase <= 60:
        return "Beban Ringan"
    elif persentase <= 85:
        return "Beban Sedang"
    else:
        return "Beban Berat"



def monitor_suhu(suhu_list):
    for i in range(len(suhu_list)):
        suhu = suhu_list[i]
        print(f"Titik {i+1}: {suhu}°C")

        if suhu > 100:
            print("Peringatan: Suhu melewati batas aman")



def pemanasan_awal():
    suhu = 60
    print("\nMemulai proses pemanasan awal...")

    while suhu < 85:
        print("Suhu saat ini:", suhu)
        suhu += 5

    print("Target 85°C tercapai\n")



def main():
    print("=== Sistem Monitoring Energi Boiler ===")


    pemanasan_awal()


    suhu_boiler = [92.5, 88.7, 95.1, 102.3, 97.8]
    print("Monitoring suhu beberapa titik:")
    monitor_suhu(suhu_boiler)


    kapasitas = 500
    terpakai = 380
    status_beban = klasifikasi_beban(kapasitas, terpakai)
    print("\nStatus beban:", status_beban)


    efisiensi = hitung_efisiensi(1000, 820)
    print(f"Efisiensi sistem: {efisiensi*100:.2f}%")


main()
```


• Program punya beberapa fungsi biar rapi dan modular.

• Loop while dipakai buat simulasi naiknya suhu awal.

• Loop for dipakai buat ngecek beberapa titik suhu di boiler.

• If-elif-else dipakai buat nentuin kategori beban.

• Ada fungsi efisiensi buat hitung performa energi.
